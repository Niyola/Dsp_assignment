In the first lecture we showed (for DTFT) that, for a *real signal*, the spectrum is symmetric, i.e.
\begin{equation}
X(-\omega)=\overline{X(\omega)}
\end{equation}

In general, we have 

1) Real signal <=> even symmetric ($X(-\omega)=\overline{X(\omega)}$) spectrum  \
2) Imaginary signal <=> odd symmetric ($X(-\omega)=-\overline{X(\omega)}$) spectrum \
3) Even symmetric signal <=> Real spectrum \
4) Odd symmetric signal <=> Imaginary spectrum 

Similar results holds for DFT with caveat that the signal must be assumed to be periodic with period $N$. For e.g., an even symmetric signal this means that $x[n]=\overline{x[-n]}=\overline{x[N-n]}$  

 ## a 
#### Show the third statement for DTFT and DFT.
 #### Let's assume we have a discrete-time signal x[n] that is even symmetric, meaning:
 
#### x[n] = x[-n]
#### Now, let's compute the DTFT of this signal, denoted as X(e^jω):
#### X(e^jω) = Σ[x[n] * e^(-jωn)] over all n
#### Since x[n] is even symmetric, we can rewrite this as:
#### X(e^jω) = Σ[x[-n] * e^(-jωn)] over all n
#### Now, let's consider the complex conjugate of X(e^jω), denoted as X*(-e^jω):
#### X*(-e^jω) = Σ[x[-n] * e^(jωn)] over all n
#### Notice that we've simply changed the sign of the exponent in the exponential term while keeping x[-n]. Since x[n] = x[-n], these two expressions are equal:
#### X(e^jω) = X*(-e^jω)
#### Therefore, for an even symmetric signal x[n], its DTFT X(e^jω) is equal to its complex conjugate X*(-e^jω), which means the spectrum is real.

## b
#### What about a complex signal? What is the intuition?
 #### To show the same property for DFT, we need to consider that the signal must be assumed to be periodic with period N. For an even symmetric signal, this means:
 
#### x[n] = x[-n] = x[N - n]
#### Now, let's compute the DFT of this signal, denoted as X[k]:
#### X[k] = Σ[x[n] * e^(-j(2π/N)kn)] over all n
#### Since x[n] is even symmetric, we can rewrite this as:
#### X[k] = Σ[x[-n] * e^(-j(2π/N)kn)] over all n
#### Now, let's consider the complex conjugate of X[k], denoted as X*(-k):
#### X*(-k) = Σ[x[-n] * e^(j(2π/N)kn)] over all n
#### Again, notice that we've simply changed the sign of the exponent in the exponential term while keeping x[-n]. Since x[n] = x[-n], these two expressions are equal:
#### X[k] = X*(-k)Therefore, for an even symmetric signal x[n] assumed to be periodic with period N, its DFT X[k] is equal to its complex conjugate X*(-k), which means the spectrum is real.
#### So, we've shown that for both DTFT and DFT, an even symmetric signal in the time domain has a real spectrum in the frequency domain.


## 2

#### To prove the inversion formula for the case of N = 2, which is essentially showing that the left-hand side (LHS) of the second equation equals the right-hand side (RHS), you'll need to calculate some values of X[k] using the first equation and then verify the equality. Let's go step by step:
#### Given:
#### DFT Equation: X[k] = Σ (n=0 to N-1) x[n] * e^(-j2πkn/N), k ∈ [0, N-1]
#### Inverse DFT Equation: x[n] = (1/N) * Σ (k=0 to N-1) X[k] * e^(j2πkn/N), n ∈ [0, N-1]
#### In this case, N = 2.


## Step 1: Calculate X[k] using the DFT equation for k = 0 and k = 1.

#### For k = 0: X[0] = Σ (n=0 to 1) x[n] * e^(-j2πkn/2) = x[0] * e^0 + x[1] * e^(-jπ) = x[0] + x[1] * (-1) # Since e^(-jπ) = -1
#### For k = 1: X[1] = Σ (n=0 to 1) x[n] * e^(-j2πkn/2) = x[0] * e^0 + x[1] * e^(-j2π/2) = x[0] + x[1] * (-j) # Since e^(-j2π/2) = -j

## Step 2: Substitute these values into the inverse DFT equation for n = 0 and n = 1.

#### For n = 0: x[0] = (1/2) * [X[0] * e^(j2π00/2) + X[1] * e^(j2π01/2)] = (1/2) * [X[0] * e^0 + X[1] * e^0] = (1/2) * [X[0] + X[1]]
#### For n = 1: x[1] = (1/2) * [X[0] * e^(j2π10/2) + X[1] * e^(j2π11/2)] = (1/2) * [X[0] * e^0 + X[1] * e^(jπ)] = (1/2) * [X[0] + X[1] * (-1)] # Since e^(jπ) = -1

## Step 3: Substitute the expressions for x[0] and x[1] back into the original sequence representation:

#### x[0] = (1/2) * [X[0] + X[1]] x[1] = (1/2) * [X[0] - X[1]]
#### These equations are the inverse DFT equations for N = 2. They show that the original sequence x[n] can be reconstructed from its DFT coefficients X[k] for N = 2 using these formulas. This verifies the inversion formula for N = 2.
